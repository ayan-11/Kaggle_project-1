```python
train=data.iloc[:,:-1]
test=data.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(train,test,train_size=0.8,random_state=42)
```

```python
from sklearn.preprocessing import OrdinalEncoder
def ord_val(data):
    enc=OrdinalEncoder(dtype='int64')
    cols=['GENDER','Car_Owner','Propert_Owner','Type_Income','EDUCATION','Marital_status','Housing_type','Type_Occupation']
    data[cols]=enc.fit(data[cols])
    return data
x_train=ord_val(x_train)
x_test=ord_val(x_test)
```
